```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_hun = df[df['job_country'] == 'Hungary'].copy()
```

```python
df_hun
```

```python
df_hun['job_posted_month'] = df_hun['job_posted_date'].dt.strftime('%B')

df_hun
```

```python
df_hun.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_hun_pivot = df_hun.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_hun_pivot
```

```python
df_hun_pivot.reset_index(inplace=True)
df_hun_pivot['month_no'] = pd.to_datetime(df_hun_pivot['job_posted_month'], format='%B').dt.month

df_hun_pivot['month_no']
```

```python
df_hun_pivot.sort_values('month_no', inplace=True)
df_hun_pivot.set_index('job_posted_month', inplace=True)
df_hun_pivot.drop(columns='month_no', inplace=True)
```

```python
df_hun_pivot
```

```python
df_hun_pivot.plot(kind='line')
```

```python
top_3 = df_hun['job_title_short'].value_counts().head(3)
```

```python
top_3 = top_3.index.to_list()
```

```python
df_hun_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Ppostings for Top Data Jobs in Hungary')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show
```
