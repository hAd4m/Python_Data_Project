Data Filtering Example

```python
job_data = [
    {"job_title": "Senior Data Scientist", "job_skills": ["Python", "SQL", "PyTorch", "Machine Learning"], "remote": True},
    {"job_title": "Data Analyst", "job_skills": ["R", "Tableau", "Excel", "Statistical Analysis"], "remote": False},
    {"job_title": "Machine Learning Engineer", "job_skills": ["Python", "TensorFlow", "Docker", "AWS", "CI/CD"], "remote": True},
    {"job_title": "Junior Data Scientist", "job_skills": ["Python", "Pandas", "Scikit-Learn", "Matplotlib"], "remote": False},
    {"job_title": "NLP Specialist", "job_skills": ["Python", "HuggingFace", "BERT", "Deep Learning"], "remote": True}
]
```

```python
filter(lambda job: job["remote"], job_data)
```

```python
list(filter(lambda job: job["remote"], job_data))
```

```python
list(filter(lambda job: job["remote"] and "Python" in job["job_skills"], job_data))
```
