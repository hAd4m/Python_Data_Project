```python
years_experience = {
    'Adam':   3,
    'AdamII': 6,
    'AdamIII':4,
    'Adam0':  3,
    'AdamIV': 7
}
```

```python
years = 0
index = 0

years_list = list(years_experience.items())

while years < 5:
  key, years = years_list[index]
  if years >= 5:
    print(key, years)
  index += 1
```
